```python
import shutil
import os 

source = '..\datasets\ISIC2017\ISIC-2017_Training_Data\\'
target = '..\datasets\ISIC2017\Training_data_processed_2\\not_successfully_cutted\\'
to_replace = os.listdir(target)

for file in to_replace:
        if file.find('superpixels') == -1 and file.find('Training_Data_metadata') == -1 and file.find('cut') == -1:
            try:
                img = file.split('_final')[0]
                shutil.copyfile(source+img+'.jpg', target+img+'.jpg')
                os.remove(target+file)
            except Exception as e:
                print(file, e)
```
